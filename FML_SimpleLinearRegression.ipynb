**importing** **required** **modules**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```


Read the csv file using pandas

```python
data=pd.read_csv("Company_data.csv")
data.head()
```

**Finding** **correaltion** **between** **the** **variables**

```python
a=data.corr()
a
```

**Here x contains all the columns except Sales and y contains the Column Sales**

```python
X=data.drop('Sales',axis=1).values
y=data['Sales'].values
y
```

**Taking TV column array from X and storing it in a variable x_tv as numpy array**

```python
x_tv = X[:, 0]
x_tv

```

```python
plt.scatter(x_tv, y)
plt.xlabel('Number of TV advertisements')
plt.ylabel('Value of sales')
plt.show()
```

**Reshaping both x_tv and y into 1_Dimensional**

```python
x_tv=x_tv.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
x_tv
```

```python
y
```

**Splitting the data into test and train**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_tv, y, test_size=0.2, random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred1 = reg.predict(X_train)
print("Performance of the model on training data :\n")
print("MAE =", mean_absolute_error(y_train, y_pred1))
print("MSE =",mean_squared_error(y_train, y_pred1))
print("RMSE =", np.sqrt(mean_squared_error(y_train, y_pred1)))
print("R_2 =",r2_score(y_train, y_pred1))
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred2=reg.predict(X_test)
print("Performance of the model on testing data :\n")
print("MAE =", mean_absolute_error(y_test, y_pred2))
print("MSE =",mean_squared_error(y_test, y_pred2))
print("RMSE =", np.sqrt(mean_squared_error(y_test, y_pred2)))
print("R_2 =",r2_score(y_test, y_pred2))
```
