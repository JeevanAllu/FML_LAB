```python
import numpy as np;
a=np.array([1,2,4])
print(type(a))
print(a.shape)
print(a.ndim)
print(a)
```

```python
b=np.array([[1,2,3],[4,5,6]])
print(b.shape)
print(b.ndim)
print(b)
```
