**Sigmoid**

```python
import numpy as np
import matplotlib.pyplot as plt
def Sigmoid(x):
  return  1/(1+np.exp(-x))
x=np.linspace(-10,10)
x
x.shape
```

```python
plt.plot(x,Sigmoid(x))
#plt.axis('tight')
plt.title("Sigmoid-Activation Function")
plt.show()
```

**tanh**:

```python
import numpy as np
import matplotlib.pyplot as plt
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('tanh-Activation Function')
plt.show()
```

**ReLU**:

```python
import numpy as np
import matplotlib.pyplot as plt
def Relu(x):
  y=[]
  for i in x:
    if(i<0):
      y.append(0)
    else:
      y.append(i)
  return y
x=np.linspace(-10,10)
plt.plot(x,Relu(x))
plt.axis('tight')
plt.title("RELU-Activation Function")
plt.show()
```

**SoftMax**-**Activation** **Functioin**

```python
import numpy as np
import matplotlib.pyplot as plt
def softmax(x):
  return np.exp(x)/(np.sum(np.exp(x),axis=0))
x=np.linspace(-10,10)
plt.plot(x,softmax(x))
plt.axis('tight')
plt.title('SOFTMAX-ACTIVATION FUNCTION')
plt.show()
```
